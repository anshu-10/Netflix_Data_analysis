```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('netflix_titles.csv')
print(df.shape)
print(df.head())
```

```python
print(df.isnull().sum())
df.dropna(inplace=True)
print(df.shape)
```

```python
df['type'] = df['type'].dropna()
type_counts = df['type'].value_counts()
type_counts.plot(kind='bar', color = ['skyblue', 'salmon'][:len(type_counts)])
plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Type")
plt.ylabel("Count")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year


df['year_added'].value_counts().sort_index().plot(kind='line', marker='o', color='purple')
plt.title("Number of Titles Added to Netflix Each Year")
plt.xlabel("Year Added")
plt.ylabel("Number of Titles (items)")
plt.grid()
plt.show()  
```

```python
df['country'].value_counts().head(10).plot(kind='bar', color='red')
plt.title("Top 10 Countries with Most Titles on Netflix")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()  
```

```python
df_filtered = df[(df['release_year'] >= 2016) & (df['release_year'] <= 2022)]

df_filtered['release_year'].value_counts().plot(kind='pie')

plt.title('Number of Titles Released (20162022)')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
